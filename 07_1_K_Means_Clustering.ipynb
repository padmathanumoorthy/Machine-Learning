```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=400, n_features=2, centers=8, cluster_std=1.8, random_state=42)
```

#### Unsupervised Learning

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=8)
kmeans
```

```python
kmeans.fit(data[0])
kmeans
```

```python
kmeans.labels_
```

```python
data[1]
```

```python
f , (axis1,axis2) = plt.subplots(1, 2, sharey=True, figsize=(10,6))

axis1.set_title('Original')
axis1.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')

axis2.set_title('K Means')
axis2.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')
```
