```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
kmeans
```

```python
df.drop('Private', axis=1) # Drop the col named 'Private'
```

```python
kmeans.fit(df.drop('Private', axis=1))
kmeans
```

```python
df
```

```python
kmeans.labels_
```

```python
def convert_to_yesno(private):
    if private == 'Yes':
        return 1
    else:
        return 0
```

```python
df['is_private'] = df['Private'].apply(convert_to_yesno)
df.tail(10)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
confusion_matrix(df['is_private'], kmeans.labels_)
```

```python
print(classification_report(df['is_private'], kmeans.labels_))
```
